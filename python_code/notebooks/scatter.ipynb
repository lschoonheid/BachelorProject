```python
all_particles_list = [e[2] for e in event_list]
all_count_list = []
for i, particles in enumerate(all_particles_list):
    particles['event_id'] = i
    all_count_list.append(len(particles))
all_particles_frame = pd.concat(all_particles_list)
print(all_count_list)
all_particles_frame
```

```python
p_r = 30
p_z = 25

all_primary_particles_list = []
primary_count_list = []
for particles in all_particles_list:
    primary_particles = particles[
    (particles['vx'] > -p_r) & (particles['vx'] < p_r) &
    (particles['vy'] > -p_r) & (particles['vy'] < p_r) &
    (particles['vz'] > -p_z) & (particles['vz'] < p_z) ]
    all_primary_particles_list.append(primary_particles)
    primary_count_list.append(len(primary_particles))
all_primary_particles = pd.concat(all_primary_particles_list)
print(primary_count_list)
print(sum(primary_count_list))
```

```python
p_r = 30
p_z = 25

all_secondary_particles_list = []
secondary_count_list = []
for particles in all_particles_list:
    secondary_particles = particles[
    (particles['vx'] < -p_r) | (particles['vx'] > p_r) &
    (particles['vy'] < -p_r) | (particles['vy'] > p_r) &
    (particles['vz'] < -p_z) | (particles['vz'] > p_z) ]
    all_secondary_particles_list.append(secondary_particles)
    secondary_count_list.append(len(secondary_particles))
all_secondary_particles = pd.concat(all_secondary_particles_list)
print(secondary_count_list)
print(sum(secondary_count_list))
```

```python
print('primary particles per event', np.mean(primary_count_list), "pm", np.std(primary_count_list)/np.sqrt(len(primary_count_list)))
print('secondary particles per event', np.mean(secondary_count_list), "pm", np.std(secondary_count_list)/np.sqrt(len(secondary_count_list)))
```
